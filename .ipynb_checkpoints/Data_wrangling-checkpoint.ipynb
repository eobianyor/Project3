# IMPORT DATA

```python
# File path for the CSV Files
idahoFires = os.path.join("Data", "fires_Idaho.csv")

# Open the CSV Files, Convert to a Dataframe, and Save as a Variable
fires_Idaho_df = pd.read_csv(idahoFires)
# fires_Idaho_df = pd.read_csv(idahoFires, dtype={"LOCAL_INCIDENT_ID": 'string', "FIRE_NAME": 'string'})
```

```python
# Convert date to datetime format
fires_Idaho_df['DISCOVERY_DATE_CONVERTED'] = pd.to_datetime(fires_Idaho_df['DISCOVERY_DATE_CONVERTED'])
fires_Idaho_df['CONT_DATE_CONVERTED'] = pd.to_datetime(fires_Idaho_df['CONT_DATE_CONVERTED'])

# View the Data in the Dataframe
# print(fires_Idaho_df.keys())
fires_Idaho_df
```

### Generate and associate cities using the lat lng coordinates

```python
from citipy import citipy

# Pull lat lng columns from df
location_df = fires_Idaho_df[['LATITUDE', 'LONGITUDE']]
```

```python
### Generate a random cities from Lat Long table make sure we have up to 500. Delete any duplicates: 
# Make lists needed
cityList = []
countryList = []

# Create a loop to find a city for each lat long pair
for index, row in location_df.iterrows():
#     print(f"Searching for nearest city to lat long {row['Latitude'], row['Longitude']}...")
    city = citipy.nearest_city(row["LATITUDE"], row["LONGITUDE"])
    cityName = city.city_name
    location_df.loc[index,"CITY"] = cityName
    fires_Idaho_df.loc[index,"CITY"] = cityName
    cityList.append(cityName)
```

```python
fires_Idaho_df
```

## DATA CLEAN UP AND REMOVE UNWANTED COLUMNS

```python
# CREATE COLUMNS WE NEED
fires_Idaho_df['FIRE_DAYS'] = (((fires_Idaho_df['CONT_DOY']) + 1) - fires_Idaho_df['DISCOVERY_DOY'])
fires_Idaho_df['COUNTY_NAME'] = (fires_Idaho_df['FIPS_NAME'])

# DELETE ROWS WE DO NOT NEED
fires_Idaho_2000_2015_df = fires_Idaho_df.loc[(fires_Idaho_df["FIRE_YEAR"] >= 1999)] 

# DELETE COLUMNS WE DO NOT NEED
fires_Idaho_XY_df = fires_Idaho_2000_2015_df.drop(['OBJECTID', 'FOD_ID', 'FPA_ID', 'SOURCE_SYSTEM_TYPE', 'SOURCE_SYSTEM', 'NWCG_REPORTING_AGENCY', 'NWCG_REPORTING_UNIT_ID', 
                                                   'NWCG_REPORTING_UNIT_NAME', 'SOURCE_REPORTING_UNIT', 'SOURCE_REPORTING_UNIT_NAME', 'LOCAL_FIRE_REPORT_ID', 'LOCAL_INCIDENT_ID', 
                                                   'FIRE_CODE', 'FIRE_NAME', 'ICS_209_INCIDENT_NUMBER', 'ICS_209_NAME', 'MTBS_ID', 'MTBS_FIRE_NAME', 'COMPLEX_NAME', 'FIRE_YEAR', 
                                                   'DISCOVERY_DOY', 'DISCOVERY_TIME', 'STAT_CAUSE_CODE', 'STAT_CAUSE_DESCR', 'CONT_DOY','CONT_TIME', 'FIRE_SIZE', 'LATITUDE', 
                                                   'LONGITUDE', 'OWNER_CODE', 'OWNER_DESCR', 'STATE', 'COUNTY', 'FIPS_CODE', 'FIPS_NAME', 'Shape', 'CONT_DATE_CONVERTED', 
                                                   'FIRE_DAYS'], axis=1)

fires_Idaho_XY_df
```

```python
# Output weather dataset
output_file = os.path.join("Data", "fires_Idaho_XY_TEST.csv")
fires_Idaho_XY_df.to_csv(output_file, index = False)
```
