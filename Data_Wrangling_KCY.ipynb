```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import os
from datetime import date

# Create path to database file
database_path = os.path.join("FPA_FOD_20170508_Corrected_Extended.sqlite")
```

```python
engine = create_engine(f"sqlite:///{database_path}")
```

```python
fires_df = pd.read_sql("SELECT * FROM FiresExtCor", engine)
fires_df
```

```python
columns_dtypes = fires_df.dtypes.to_dict()
```

```python
columns_dtypes
```

```python
fires_df.dropna(subset=['LATITUDE', 'LONGITUDE','DISCOVERY_DATE','CONT_DATE', 'STATE', 'COUNTY', 'FIRE_SIZE', 'CONT_DOY'], inplace=True)
```

```python
fires_df
```

```python
# fires_subset = fires_df[fires_df['FIRE_SIZE_CLASS'].isin(['G'])].copy()
# fires_subset
```

```python
fires_subset = fires_df[fires_df['STATE'].isin(['ID'])].copy()
fires_subset
```

```python
fires_subset['DISCOVERY_DATE_CONVERTED'] = pd.to_datetime(fires_subset['DISCOVERY_DATE'], unit='D', origin='julian')
fires_subset['CONT_DATE_CONVERTED'] = pd.to_datetime(fires_subset['CONT_DATE'], unit='D', origin='julian')
fires_subset
```

```python
fires_subset = fires_subset.assign(Shape ='Point')
fires_subset
```

```python
fires_subset['CONT_DOY'] = fires_subset['CONT_DOY'].fillna(0)
```

```python
fires_subset['CONT_DOY'] = fires_subset['CONT_DOY'].astype(int)
```

```python
# fires_clean = fires_subset.drop(['OBJECTID', 'DISCOVERY_DATE', 'CONT_DATE'], axis=1)
# fires_clean

fires_clean = fires_subset.drop(['DISCOVERY_DATE', 'CONT_DATE'], axis=1)
fires_clean
```

```python
fires_clean.dtypes
```

```python
fires_YEAR = fires_clean[fires_clean['FIRE_YEAR'].isin([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015])].copy()

fires_YEAR
```

```python
drought_data = pd.read_csv("Idaho_DM_weekly_cumulative_20000101_20210120.csv")
drought_data.head()
```

```python
drought_data['County'] = drought_data['County'].str.replace(' County', '')
drought_data.head()
```

```python
drought_data.dtypes
```

```python
drought_data['ValidStart'] = pd.to_datetime(drought_data['ValidStart'])
drought_data['ValidEnd'] = pd.to_datetime(drought_data['ValidEnd'])
```

```python
drought_data.dtypes
```

```python
fires_YEAR['DM_FILENAME'] = ''
fires_YEAR['None'] = 0
fires_YEAR['D0'] = 0
fires_YEAR['D1'] = 0
fires_YEAR['D2'] = 0
fires_YEAR['D3'] = 0
fires_YEAR['D4'] = 0

fires_YEAR['None'] = fires_YEAR['None'].astype(float)
fires_YEAR['D0'] = fires_YEAR['D0'].astype(float)
fires_YEAR['D1'] = fires_YEAR['D1'].astype(float)
fires_YEAR['D2'] = fires_YEAR['D2'].astype(float)
fires_YEAR['D3'] = fires_YEAR['D3'].astype(float)
fires_YEAR['D4'] = fires_YEAR['D4'].astype(float)
```

```python
fires_YEAR.reset_index(drop=True, inplace=True)
```

```python
fires_YEAR.head()
```

```python
count = 1
for index, row in fires_YEAR.iterrows():
    county = row['NAME']
    discoDate = row['DISCOVERY_DATE_CONVERTED']
    fod_id = row['FOD_ID']
    
    print(f"Record: {count},   FOD_ID {fod_id}")
    
    found_df = drought_data[(drought_data['ValidStart'] <= discoDate) & (drought_data['ValidEnd'] >= discoDate) & (drought_data['County'] == county)]
    
    if found_df.empty:
        print("Found df is empty")
    else:
        print("Found df is not empty")  

        fires_YEAR.loc[index,'DM_FILENAME'] = found_df.iloc[0,0]
        fires_YEAR.loc[index,'None'] = found_df.iloc[0,4]
        fires_YEAR.loc[index,'D0'] = found_df.iloc[0,5]
        fires_YEAR.loc[index,'D1'] = found_df.iloc[0,6]
        fires_YEAR.loc[index,'D2'] = found_df.iloc[0,7]
        fires_YEAR.loc[index,'D3'] = found_df.iloc[0,8]
        fires_YEAR.loc[index,'D4'] = found_df.iloc[0,9]
        
    count = count + 1
    
    
    
```

```python
fire_weather_data = pd.read_csv("Idaho_Fire_Weather_Data.csv")
```

```python
fire_weather_data.head()
```

```python
fire_weather_data['avg_temp'] = fire_weather_data['avg_temp'] * 1.8 + 32
```

```python
fire_weather_data.head()
```

```python
len(fire_weather_data)
```

```python
cols = ['FOD_ID', 'DAY1','DAY2','DAY3','DAY4','DAY_PRCP_1','DAY_PRCP_2','DAY_PRCP_3','DAY_PRCP_4','DAY_AVG_TEMP_1','DAY_AVG_TEMP_2','DAY_AVG_TEMP_3','DAY_AVG_TEMP_4']
```

```python
weather_flipped = pd.DataFrame(columns=cols)
```

```python
weather_flipped.head()
```

```python
unique_Ids = fire_weather_data.FOD_ID.unique().tolist()
```

```python
unique_Ids
```

```python
for id in unique_Ids:
    print(id)
    weather_id = fire_weather_data[(fire_weather_data['FOD_ID'] == id)]
    
    if(len(weather_id)==4):
        weather_flipped = weather_flipped.append({
        'FOD_ID' : id , 
        'DAY1' : weather_id.iloc[0,3],
        'DAY2' : weather_id.iloc[1,3],
        'DAY3' : weather_id.iloc[2,3],
        'DAY4' : weather_id.iloc[3,3],
        'DAY_PRCP_1' : weather_id.iloc[0,2],
        'DAY_PRCP_2' : weather_id.iloc[1,2],
        'DAY_PRCP_3' : weather_id.iloc[2,2],
        'DAY_PRCP_4' : weather_id.iloc[3,2],
        'DAY_AVG_TEMP_1' : weather_id.iloc[0,1],
        'DAY_AVG_TEMP_2' : weather_id.iloc[1,1],
        'DAY_AVG_TEMP_3' : weather_id.iloc[2,1],
        'DAY_AVG_TEMP_4' : weather_id.iloc[3,1]
        }, ignore_index=True)
    elif(len(weather_id)==3):
        weather_flipped = weather_flipped.append({
        'FOD_ID' : id , 
        'DAY1' : weather_id.iloc[0,3],
        'DAY2' : weather_id.iloc[1,3],
        'DAY3' : weather_id.iloc[2,3],
        'DAY_PRCP_1' : weather_id.iloc[0,2],
        'DAY_PRCP_2' : weather_id.iloc[1,2],
        'DAY_PRCP_3' : weather_id.iloc[2,2],
        'DAY_AVG_TEMP_1' : weather_id.iloc[0,1],
        'DAY_AVG_TEMP_2' : weather_id.iloc[1,1],
        'DAY_AVG_TEMP_3' : weather_id.iloc[2,1]
        }, ignore_index=True)
        
    elif(len(weather_id)==2):
        weather_flipped = weather_flipped.append({
        'FOD_ID' : id , 
        'DAY1' : weather_id.iloc[0,3],
        'DAY2' : weather_id.iloc[1,3],
        'DAY_PRCP_1' : weather_id.iloc[0,2],
        'DAY_PRCP_2' : weather_id.iloc[1,2],
        'DAY_AVG_TEMP_1' : weather_id.iloc[0,1],
        'DAY_AVG_TEMP_2' : weather_id.iloc[1,1]
        }, ignore_index=True)
    else:
        weather_flipped = weather_flipped.append({
        'FOD_ID' : id , 
        'DAY1' : weather_id.iloc[0,3],
        'DAY_PRCP_1' : weather_id.iloc[0,2],
        'DAY_AVG_TEMP_1' : weather_id.iloc[0,1],
        }, ignore_index=True)
    
weather_flipped.head()   
```

```python
fires_w_drought_weather = pd.merge(fires_YEAR, weather_flipped, how='left', on='FOD_ID')
```
